```python

data2020url = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2020/COVID19MEXICO2020.zip"

data2021url = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2021/COVID19MEXICO2021.zip"

data2022url = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2022/COVID19MEXICO2022.zip"

data2023url = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2023/COVID19MEXICO2023.zip"

data2024url = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2024/COVID19MEXICO2024.zip"

data2025url = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/historicos/2025/12/datos_abiertos_influenza_covid19_30.12.2025.zip"

data2026url = "https://datosabiertos.salud.gob.mx/gobmx/salud/datos_abiertos/datos_abiertos_influenza_covid19.zip"

datadir = [data2020url, data2021url, data2022url, data2023url, data2024url, data2025url, data2026url]
```

```python
from pathlib import Path
import re

def rename_files(datadir):
  """
  Function to normalize names of files to be downloaded
  """
  normalized = []

  for url in datadir:

    name = Path(url).name
    year = re.search(r"(20\d{2})", name)

    if year:
      year = year.group(1)
    else:
      year = "2026"

    new_name = f"DATA_RAW_{year}.zip"

    normalized.append(new_name)

  return normalized
```

```python
print(rename_files(datadir))
names = rename_files(datadir)
for url, location in zip(datadir, names):
  print(url, location)
```

```python
import requests
from pathlib import Path

names = rename_files(datadir)

ROOT = Path.cwd().parent
DATA_DIR = ROOT / "data"

for url, location in zip(datadir, names):

  filepath = DATA_DIR / location
  filepath.parent.mkdir(parents=True, exist_ok=True)

  print(f"Downloading {filepath.name}")

  with requests.get(url, stream=True) as response:
    print(f"Status code: {response.status_code}")
    response.raise_for_status()

    with filepath.open("wb") as file:
      print(f"Writing to {filepath.name}")
      for chunk in response.iter_content(chunk_size=1024 * 1024):
        if chunk:
          file.write(chunk)

  print(f"File saved at {filepath}")
```

```python
import zipfile

zip_files = [DATA_DIR / name for name in names]

for zip_file in zip_files:
    try:
        with zipfile.ZipFile(zip_file, 'r') as zf:
            print(f"'{zip_file}' is a valid zip file. It contains: {zf.namelist()}")
    except zipfile.BadZipFile:
        print(f"'{zip_file}' is NOT a valid zip file and might be corrupted. Please re-download it.")
    except FileNotFoundError:
        print(f"'{zip_file}' not found.")
```
