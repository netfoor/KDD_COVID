```python
from dotenv import load_dotenv
from pymongo import MongoClient
import os


load_dotenv()

MONGODB_URI = os.getenv("MONGO_URI")
client = MongoClient(MONGODB_URI)
db = client['covid']

collections = sorted(db.list_collection_names())

mexico_data = []

for collection in collections:
        year = collection.split("_")[1]
        count = db[collection].count_documents({"ENTIDAD_RES": "09"})
        mexico_data.append({int(year): count})

```

```python
print(mexico_data)

cases_count = 0

for cases in mexico_data:
    cases_count += cases[list(cases.keys())[0]]

print(f"Total cases in Mexico: {cases_count}")
```

```python
aggregate = db.raw_2020_data.aggregate([
    {
        "$group": {
            "_id": "$ENTIDAD_RES",
            "count": {"$sum": 1}
        }
    },
    {"$sort": {"count": -1}}
])

print(list(aggregate))
```
